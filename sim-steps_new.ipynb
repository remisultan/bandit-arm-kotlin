```kotlin
%use lets-plot
```

```kotlin
data class SimStep(val sim:Int, 
                   val horizon:Int, 
                   val armNumber:Int, 
                   val reward:Float, 
                   val cumulativeReward:Float)
```

```kotlin
val simSteps = java.io.File("/tmp/epsilson-annealed.tsv")
    .readLines()
    .map { it.split("\t") }
    .map {SimStep(it[0].toInt(), it[1].toInt(), it[2].toInt(), it[3].toFloat(), it[4].toFloat())}
```

```kotlin
val avgRewards = simSteps.groupBy { it.horizon }
val x = avgRewards.keys.toList()
val y = avgRewards.values.map{ sim -> sim.map{ it.reward }.average() }
val y1 = avgRewards.values.map{ sim -> sim.map{ it.cumulativeReward }.average() }
val y1Max = y1.max()
var plot = lets_plot(mapOf<String, Any?>(
    "horizon" to x,
    "rewards" to y,
    "cumulative" to y1.map { it / y1Max!! }
))

plot += geom_line{x="horizon"; y="rewards";alpha=0.1}
plot += geom_line(color="red"){x="horizon"; y="cumulative";alpha=0.1}
plot + ggsize(1000, 500)
```
